```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# функция расчёта для метода итераций

# phi - 𝜑(𝑥) [дальше такое же обозначение, только для 1-ой функции phi1, для второй phi2]
# dphi - 𝜑'(𝑥) [здесь не используется, но в дальнейшем записывается также dphi1и dphi2 для 1-ой и 2-ой функции]
# x0 - начальное приближение
# eps - ε - точность
# max_iterations - максимальное количество итераций, нужно чтоб если не нашли корня, то не зацикливались и программа завершилась
def iteration_method(phi, x0, eps, max_iterations=1000):
    x_prev = x0
    for n in range(1, max_iterations + 1):
        x_next = phi(x_prev)
        if abs(x_next - x_prev) < eps:
            return x_next, n
        x_prev = x_next
    return x_prev, max_iterations
```

```python
# Вывод графика первой функции

def f1(x):
    return 2**(-x) - np.sin(x)

x = np.linspace(0, 5, 1000)

plt.figure(figsize=(8, 5))
plt.plot(x, f1(x), label=r'$f(x)=2^{-x}-\sin x$')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title(r'$f(x)=2^{-x}-\sin x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
phi1 = lambda x: math.asin(2**(-x))
dphi1 = lambda x: abs(-(2**(-x) * math.log(2)) / math.sqrt(1 - 2**(-2*x)))

a1, b1 = 0.6, 0.7
q1 = max(dphi1(a1), dphi1(b1))

print(f'Интервал: [{a1}; {b1}]')
print('𝜑(x) = arcsin(2^(-x))')
print(f'max |𝜑\'(x)| = {q1:.6f}')
```

```python
x1, n1 = iteration_method(phi1, 0.7, 0.0001)

print(f'Приближённый корень: x = {x1:.6f}')
print(f'Проверка f(x) = {f1(x1):.6f}')
print(f'Число итераций: {n1}')
```

```python
# Вывод графика второй функции

def f2(x):
    return x**3 - x**2 + 7

x = np.linspace(-4, 3, 1000)

plt.figure(figsize=(8, 5))
plt.plot(x, f2(x), label=r'$f(x)=x^3-x^2+7$')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title(r'$f(x)=x^3-x^2+7$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
phi2 = lambda x: np.cbrt(x**2 - 7)
dphi2 = lambda x: abs((2 * x) / (3 * np.abs(x**2 - 7)**(2/3)))

a2, b2 = -2, -1
grid = np.linspace(a2, b2, 1000)
q2 = np.max(dphi2(grid))

print(f'Интервал: [{a2}; {b2}]')
print('𝜑(x) = (x^2 - 7)^(1/3)')
print(f'max |𝜑\'(x)| = {q2:.6f}')
```

```python
x2, n2 = iteration_method(phi2, -2, 0.01)

print(f'Приближённый корень: x = {x2:.6f}')
print(f'Проверка f(x) = {f2(x2):.6f}')
print(f'Число итераций: {n2}')
```

```python
print('Ответ:')
print(f'1) x ≈ {x1:.6f}')
print(f'2) x ≈ {x2:.6f}')
```
